# Q) What is the number of purchases with negative profits for each segment?

```python
data['Segment'].unique() # Shows the unique entries of the column "Segment"
```

```python
low_profit=data[data['Profit']<0] 
plot_low=low_profit['Segment'].value_counts() # Counts the value for the given condition
```

```python
ax=plot_low.plot(kind='bar', color='red') # Used to plot the bar graph
plt.title("Chart for Negative profit Segment")
plt.ylabel("Number of purchases")
plt.xlabel("Name of Segment")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # To rotate the Xlabels by 45 degree
```

# Q) What is the number of purchases with profits for each segment?

```python
high_profit=data[data['Profit']>0]
plot_high=high_profit['Segment'].value_counts()
```

```python
ax1=plot_high.plot(kind='bar')
plt.title("Chart for Positive Profit Segment")
plt.ylabel("Number of purchases")
plt.xlabel("Name of Segment")
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
```

Here we could observe that the most profitable segment is Consumer then Corporate and at least is Home Office.

# Q) What is the overall profit for each segment?

```python
consumer_segment=data[data['Segment']=='Consumer'] 
a1=consumer_segment[['Profit']].sum(axis=0)         # Adds the profit for the data where "Segment" is Consumer
```

```python
corporate_segment=data[data['Segment']=='Corporate']
a2=corporate_segment[['Profit']].sum(axis=0)         # Adds the profit for the data where "Segment" is Corporate
```

```python
Homeoffice_segment=data[data['Segment']=='Home Office']
a3=Homeoffice_segment[['Profit']].sum(axis=0)          # Adds the profit for the data where "Segment" is Home Office
```

```python
frames=[a1,a2,a3]
bars=['Consumer','Customer','Home Office']
seg=pd.concat(frames)
ax1=seg.plot(kind='bar', color='blue')
plt.title("Profit for each segment")
y_pos = np.arange(len(bars))
plt.xticks(y_pos,bars)
plt.ylabel("Profit")
plt.xlabel("Segment Name")

for i in ax1.patches:                                       # This loop is used for the anotation in the bar graph
    ax1.text(i.get_x(), i.get_height()+ 0.4,
    str(round(i.get_height(), 2)), fontsize=10, color='red')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
```

Hemce Consumer has the largest share in the profit for this dataset.

<br><br><br>




Therefore, by this it is clear that the company should focus more on comsumer segment as it's providing greater profits but with that it also has to improve it's presence in the Home Office segment also so that they could catch up with this segment as in the orther two segment.

# Q) Which category has the maximum profit?

```python
data['Category'].unique()
```

```python
os_category=data[data['Category']=='Office Supplies']
b1= os_category[['Profit']].sum(axis=0)                 # Adds the profit for the data where "Category" is Office Supply
```

```python
furniture_category=data[data['Category']=='Furniture']
b2= furniture_category[['Profit']].sum(axis=0)          # Adds the profit for the data where "Category" is Furniture
```

```python
tech_category=data[data['Category']=='Technology']
b3= tech_category[['Profit']].sum(axis=0)           # Adds the profit for the data where "Category" is Technology
```

```python
frames=[b1,b2,b3]
bars=['Office Supplies','Furniture','Technology']
seg=pd.concat(frames)
ax2=seg.plot(kind='barh', color='brown')
plt.title("Profit for each category")
x_pos = np.arange(len(bars))
plt.yticks(x_pos,bars)
plt.xlabel("Profit")
plt.ylabel("Category Name")
plt.grid(True)

for j in ax2.patches:                                           # This loop is used for the anotation in the bar graph
    ax2.text(j.get_width()+.2, j.get_y()+ 0.2,
    str(round(j.get_width(), 2)), fontsize=10, color='black')
```

This horizontal bar graph clearly shows that the company should consider the Furniture category and work on itproperly so as to match the profit margin with Technology and Office Supplies as it has very less profit than the other two categories.

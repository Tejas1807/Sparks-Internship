# Ploting a Running Bar graph for the number of matches won by each team.

```python
dict_season={}
for season in new_data['season'].unique():                                      # Dividing the win for each team according to the season and storing as a dictionary
    season_wins=new_data[new_data['season']==season]['winner'].value_counts()
    dict_season[season]=dict(season_wins)
print(dict_season)
```

```python
s=pd.DataFrame(dict_season)         # Converting the dictionary into a dataframe
s
```

```python
s1=s.fillna(0)      # filling all the NaN value with 0 as it would help in further exploration
```

```python
s1.head(15)
```

```python
se=new_data['season'].unique()      # storing different season in a list
se
```

```python
for i in range(12):                 # loop where i goes from0 to 11
    s1[i]=s1[se[0:i+1]].sum(axis=1) # Calculating the total number of wins till that season
s1.head()
```

```python
sum_season=s1[range(12)]        # Storing the data in which we are calculating the total number of wins till that season in sum_season
sum_season.head()
```

```python
for s in range(12):
    sum_season.rename(columns={s:se[s]}, inplace=True)          # Renaming the columns to the year till which the season has gone
sum_season
```

```python
sum_new=sum_season.reset_index()                        # Reseting the index of the dataset
sum_new.rename(columns={'index':'team'}, inplace=True)
sum_new
```

```python
int(sum_new[2009][sum_new['team']=='Rajasthan Royals'])         # This is how we can get a value from this dataset
```

```python
# bringing the data into the correct format where we could use it to plot the running bar graph
team=[]

for x in sum_new['team'].unique():
    i=0
    while i<12:
        team.append(x)
        i+=1

year=[]
i=0
while i<15:
    for x in sum_new.columns[1:]:
    
            year.append(x)
    i+=1       
```

```python
wins=[]
for x,y in zip(team,year):
    wins.append(int(sum_new[y][sum_new['team']==x]))
```

Now we have got all the lists i.e.(team, year, wins) we wanted. We just need to put it in the dictionary and convert it into a dataframe.

```python

df=pd.DataFrame(
                {
                "Team":team,
                "Year":year,
                "Wins":wins
                }
               )
df.head(20)
```

Now we begin to start our ploting of the running bar graph.

```python
last_year = 2019
dff = (df[df['Year'].eq(last_year)]
       .sort_values(by='Wins', ascending=True)
       .head(10))
dff
```

```python
colors = dict(zip(dff['Team'].unique(),
    ['#adb0ff', '#ffb3ff', '#90d595', '#e48381',
     '#aafbff', '#f7bb5f', '#eafb50','#C86A1C','#D4AC0D',
     '#BB8FCE','#AEB6BF','#76D7C4','#F7DC6F','#5DADE2','#F1948A']
))
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
dff = dff[::-1]   # flip values from top to bottom
# pass colors values to `color=`
ax.barh(dff['Team'], dff['Wins'], color=[colors[x] for x in dff['Team']])
# iterate over the values to plot labels and values 
for i, (value, name) in enumerate(zip(dff['Wins'], dff['Team'])):
    ax.text(value, i,     name,            ha='right')  
    ax.text(value, i,     value,           ha='left', color='white')  
# Add year right middle portion of canvas
ax.text(1, 0.4, last_year, transform=ax.transAxes, size=46, ha='right')
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
def draw_barchart(year):
    dff = df[df['Year'].eq(year)].sort_values(by='Wins', ascending=True).tail(10)
    ax.clear()
    ax.barh(dff['Team'], dff['Wins'], color=[colors[x] for x in dff['Team']])
    dx = dff['Wins'].max()/200
    for i, (value, name) in enumerate(zip(dff['Wins'], dff['Team'])):
        ax.text(value-dx, i-0.15,     name,           size=14, weight=600, ha='right', va='bottom',color='blue')     # Name of the team on the bar
        ax.text(value, i,     f'{value:,.0f}',  size=14, ha='left',  va='center', color='red')                       # Win count on right sode of the bar
    # ... polished styles
    ax.text(1, 0.4, year, transform=ax.transAxes, color='#777777', size=46, ha='right', weight=800)
    ax.text(0, 1.06, 'Matches', transform=ax.transAxes, size=12, color='#777777')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')          # Xlabel position on top
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'IPL matches won by each team from 2008 to 2019',
            transform=ax.transAxes, size=24, weight=600, ha='left', color='#DC7633')
    ax.text(1, 0, '@tejas', transform=ax.transAxes, ha='right',
            color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
    
draw_barchart(2019)
```

# Q) Probability of winning a match from super over if batting first?

```python
super_over=deliveries_data[deliveries_data['is_super_over']==1]         # Selecting the data where matches went to super over
super_over.head(10)
```

```python
matches=super_over['match_id'].unique()             # Getting the unique match_id values
matches
```

```python
d={}
win=0
loss=0
for match in matches:
    i=3
    j=super_over[super_over['match_id']==match]
    innings=j[j['inning']==i]
    ini_3=innings.total_runs.sum(axis=0)                    # 3rd innings score for a particular match
    k=4
    ini=j[j['inning']==k]
    ini_4=ini.total_runs.sum(axis=0)                        # 4th innings score for a particular match
    if ini_3 > ini_4:                                       # checking whether 3rd inning score was greater that 4th inning score
        win+=1
        d['win']=win
    else:
        loss+=1
        d['loss']=loss

print(d)
```

```python
keys = d.keys()
values = d.values()
plt.bar(keys, values, color=['orange','blue'])
```

Yes, it is more probable that you'll win if you are batting first in the super over.
